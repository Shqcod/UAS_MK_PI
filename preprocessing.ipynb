```python
import pandas as pd
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("kompas_gaza_articles.csv")

df = df.dropna(subset=["title", "body"])
```

```python
df["text"] = df["title"] + " " + df["body"]
```

```python
# Inisialisasi tools Sastrawi
stemmer = StemmerFactory().create_stemmer()
stopwords = set(StopWordRemoverFactory().get_stop_words())

def preprocess(text):
    # 1. lower
    text = text.lower()

    # 2. hanya huruf dan angka
    text = re.sub(r"[^a-z0-9\s]", " ", text)

    # 3. token
    tokens = text.split()

    # 4. hapus stopwords
    tokens = [t for t in tokens if t not in stopwords]

    # 5. stemming
    tokens = [stemmer.stem(t) for t in tokens]

    # gabungkan kembali
    return " ".join(tokens)
```

```python
# Terapkan preprocessing
df["clean_text"] = df["text"].apply(preprocess)
```

```python
output_path = "dataclean.csv"
df.to_csv(output_path, index=False, encoding="utf-8")

print(f"\nHasil preprocessing berhasil disimpan ke: {output_path}")
```
